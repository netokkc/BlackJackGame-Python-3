# BlackJack Game

```python

# IMPORTED RANDOM LIBRARY
import random 


#SETUP GLOBAL VARIABLES
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs') # TUPLE 

ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine',
         'Ten', 'Jack', 'Queen','King', 'Ace') #TUPLE

values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7,
          'Eight':8,'Nine':9, 'Ten':10, 'Jack':10,
          'Queen':10, 'King':10, 'Ace':11} #DICTIONARY OF VALUES OF BOTH TUPLES

# BOOLEAN TO HAVE GAME BEING PLAYED OR NOT. USED TO CREATE THE WHILE LOOP
playing = True 
```

```python
# CREATE A CLASS FOR THE CARDS WITH THEIR ATTRIBUTES
class Card():
    # TWO ATTRIBUTES
    def __init__(self, suit,rank):
        self.suit = suit
        self.rank = rank
    # THIS WILL REPORT BACK THE SUIT AND RANK OF A CARD
    def __str__(self):
        return self.rank + " of "+self.suit
```

```python
# CREATED THIS DECK CLASS TO REPRESENT 52 CARDS
class Deck:
    
    def __init__(self):
        self.deck = []  # start with an empty list
        for suit in suits:
            for rank in ranks:
                # APPENDING THE ATTRIBUTES TO THE EMPTY LIST TO FORM DECK
                self.deck.append(Card(suit,rank))
    
    # THIS METHOD PRINTS OUT THE DECK
    def __str__(self):
        deck_comp = ''
        for card in self.deck: # FOR EVERY CARD IN THE SELF.DECK LIST
            deck_comp += '\n'+ card.__str__() # ADD A NEW LINE AND PRINT THE REPRESENTATION OF THAT
        return "The deck has: "+deck_comp # PRINT OUT THE DECK 
# SHUFFLE THE DECK
    def shuffle(self):
        random.shuffle(self.deck)
 # METHOD TO GRAB A CARD FROM THE LIST      
    def deal(self):
        single_card = self.deck.pop() #grab the deck atritube and pop off
        return single_card 
```

```python
'''
Creating a Hand class with method to add, adjust for aces

'''

class Hand:
    def __init__(self):
        self.cards = []  # start with an empty list as we did in the Deck class
        self.value = 0   # start with zero value
        self.aces = 0    # add an attribute to keep track of aces
   
    #THIS METHOD ADDS A CARD TO PLAYERS HAND
    def add_card(self,card): # grabbing CARD from a method early
        
        # THIS CARD IS BEING PASSED FROM deck.deal() ---> single card(suit,rank)
        self.cards.append(card)
         
        #grabs the rank from global variable and value from dictionary
        # add to self.value = 0 with the value
        self.value += values[card.rank]
        
        #track ace
        if card.rank == "Ace": #IF THE RANK OF THAT CARD IS EQUAL TO ACE
            self.aces += 1
            
            
    def adjust_for_ace(self): # WE ARE CONSIDERING ACES ARE EQUAL TO 11
        
        # IF TOTAL VALUE > 21 AND I STILL HAVE AN ACE
        # THEN CHANGE MY ACE TO BE A 1 INSTEAD of AN 11
        while self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1
```

```python
'''
Creates a class for the chips and assigns a default value of 100 for a player
Different methods for winning or losing a bet that will updated on total

'''

class Chips:
    
    def __init__(self):
        self.total = 100  # This can be set to a default value or supplied by a user input
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
    
    def lose_bet(self):
        self.total -= self.bet
```

```python
def take_bet(chips): #THIS FUNCTIONS GETS AN INPUT FROM USER AND SETS TO CURRENT BET
    
    while True:
        
        try:  # BELOW ASKS FOR BET
            chips.bet = int(input("How many chips would you like to bet? "))
        except:
            print("Please provide a number")
        else:
            if chips.bet > chips.total:
                print("Sorry you do not have that many chips. You have: {}".format(chips.total))
            else:
                break
```

```python
def hit(deck,hand):
    #it takes the deck and player's hand
    
    single_card = deck.deal() # grabs a card from the deck
    hand.add_card(single_card) #add to players hand
    hand.adjust_for_ace() #checks for aces and adjusts 
```

```python
# THIS FUNCTION GETS INPUT TO HIT OR STAND

def hit_or_stand(deck,hand): 
    global playing  # to control an upcoming while loop
    
    while True:
        x = input("Hit or Stand? Enter H or S") # INPUT FROM PLAYER
        
        if x[0].lower() == 'h': #THIS BASICALLY GRABS THE FIRST LETTER 
            hit(deck,hand)      # EITHER LOWER OR UPPERCASE OR  ANY WORD
                                # THAT STARTS WITH H WILL BE CONSIDERED "HIT"
        
        
        elif x[0].lower() == 's': #SAME AS ABOVE BUT WITH S
            print("Player Stands. Dealer's Turn.")
            playing = False
            
            
        else: #IN CASE PLAYER INPUTS SOMETHING THAT DOES NOT START WITH H OR S
            print("Sorry I did not understand that. Please enter H or S only!")
            continue
        
        break
```

```python
def show_some(player,dealer):
    print("\n")
    print("DEALERS HAND:")
    print("ONE CARD HIDDEN!")
    print(dealer.cards[1])
    print("\n")
    print("PLAYERS HAND:")
    for card in player.cards:
        print(card)
    
def show_all(player,dealer):
    
    print("DEALER'S HAND:")
    for card in dealer.cards:
        print(card)
    print("\n")
    print("PLAYER'S HAND:")
    for card in player.cards:
        print(card)
```

```python
'''
Different functions for game scenarios such as BUST,WIN or PUSH for both player and dealer 
It updates the chips class total
'''

def player_busts(player,dealer,chips): #passing these from different classes
    print("PLAYER BUSTED!")
    chips.lose_bet() #calling method to adjust for total
    
def player_wins(player,dealer,chips):
    print("PLAYER WINS!")
    chips.win_bet() # calling method to adjust balance for player

def dealer_busts(player,dealer,chips):
    print("DEALER BUSTED! PLAYER WINS!")
    chips.win_bet() # calling method to adjust balance for player
    
def dealer_wins(player,dealer,chips):
    print("DEALER WINS!")
    chips.lose_bet()
    
def push(player,dealer):
    print("TIE! Push!")
```

```python
while True:
    # Print an opening statement

    print("WELCOME TO BLACKJACK GAME!")
    # Create & shuffle the deck, deal two cards to each player
    deck = Deck() # create a deck
    deck.shuffle() # shuffle the deck
    
    # SETUP THE PLAYERS HAND
    player_hand = Hand() # create a hand to player by calling method
    player_hand.add_card(deck.deal()) # deal 1st card
    player_hand.add_card(deck.deal()) #deal 2nd card
    
    # SETUP THE DEALERS HAND
    dealer_hand = Hand() #create a hand to dealer by calling method
    dealer_hand.add_card(deck.deal()) # deal 1st card
    dealer_hand.add_card(deck.deal()) #deal 2nd card
        
    # Set up the Player's chips
    player_chips = Chips()
    
    # Prompt the Player for their bet
    take_bet(player_chips)
    
    # Show cards (but keep one dealer card hidden)
    show_some(player_hand,dealer_hand) # one of dealer's cards are facind down
    
    while playing:  # recall this variable from our hit_or_stand function
        
        # Prompt for Player to Hit or Stand
        hit_or_stand(deck, player_hand)
        
        # Show cards (but keep one dealer card hidden)
        show_some(player_hand,dealer_hand)
        
        # If player's hand exceeds 21, run player_busts() and break out of loop
        if player_hand.value > 21:
            player_busts(player_hand,dealer_hand,player_chips)

            break

    # If Player hasn't busted, play Dealer's hand until Dealer reaches 17
    if player_hand.value <= 21:
        
        while dealer_hand.value < 17: #CHECKS FOR SOFT 17 RULE
            hit(deck,dealer_hand)
    
        # Show all cards
        show_all(player_hand,dealer_hand)
        
        # Run different winning scenarios
        if dealer_hand.value > 21:
            dealer_busts(player_hand,dealer_hand,player_chips)
        elif dealer_hand.value > player_hand.value:
            dealer_wins(player_hand,dealer_hand,player_chips)
        elif dealer_hand.value < player_hand.value:
            player_wins(player_hand,dealer_hand,player_chips)
        else:
            push(player_hand,dealer_hand)
    
    # Inform Player of their chips total 
    print("\n PLAYER TOTAL CHIPS ARE AT: {}".format(player_chips.total))
    
    # Ask to play again
    new_game = input("WOULD YOU LIKE TO PLAY ANOTHER HAND? Y/N")
    
    if new_game[0].lower() == 'y':
        playing = True
        continue
    else:
        print("THANK YOU FOR PLAYING!")

        break
```
